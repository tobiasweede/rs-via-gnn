```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
import re

figure_path = '/home/weiss/git/thesis/doc/figures/'
```

```python
def show_vis(name):
    file = 'results/' + name + '/logs.txt'
    try:
        with open(file) as f:
            data_list = re.findall(r"(\d\.\d+),", f.read())
            data_list =  [float(x) for x in data_list]
        train_loss = pd.DataFrame(zip(range(1,301), data_list[::2]), columns=['iter', 'rmse'])
        print(train_loss)
        val_loss = pd.DataFrame(zip(range(1,301), data_list[1::2]), columns=['iter', 'rmse'])
        print(val_loss)
        plt.rcParams['font.size'] = 26
        plt.rcParams['axes.labelsize'] = 26
        plt.rcParams['xtick.labelsize'] = 26
        plt.rcParams['ytick.labelsize'] = 26
        plt.rcParams['legend.fontsize'] = 26
        plt.plot('iter', 'rmse', data=train_loss, label='train')
        plt.plot('iter', 'rmse', data=val_loss, label='val')
        plt.xlabel("epoch")
        plt.ylabel("loss")
        plt.legend()
        plt.style.use('seaborn-whitegrid')
        plt.ylim([0.2, 1.8])
        plt.savefig(figure_path + '/' + name + '-igmc-loss.png', dpi=300, bbox_inches='tight')
        plt.show()
    except Exception as e:
        print(e)
```

```python
name='ml_100k_mnph201_testmode'
show_vis(name)
```
